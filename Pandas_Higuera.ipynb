```python
olympics = pd.read_csv('Copia de Summer-Olympic-medals-1976-to-2008.csv', sep=",", encoding ='latin-1' )

olympics.head(2)
```

```python
olympics.set_index('Year', inplace= True)
```

```python
olympics.head(2)
```

```python
nombre_completo = olympics["Athlete"].str.split(',',expand=True)
nombre_apellido = nombre_completo.drop(columns=[2])
nombre_apellido.columns = ['last_name', 'first_name']
olympics= pd.concat([olympics, nombre_apellido], axis=1)
olympics.head(3)

```

```python
city_medals = olympics.groupby(['Year', 'City', 'Medal']).count()
resumen = olympics.describe()
resumen
```

```python
city_medals
```

```python
ciudad_top = resumen.loc['top', 'City']
print(f'La ciudad que ganó más medallas fue {ciudad_top}.')
```

```python
solo_hombres = olympics.loc[olympics['Gender'] == "Men"]
```

```python
tabla = solo_hombres.groupby(['Year', 'Medal']).count()
tabla
```
